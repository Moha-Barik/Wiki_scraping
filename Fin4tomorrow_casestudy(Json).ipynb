```python
# import the requests and BeautifulSoup libraries
import requests
import bs4
from bs4 import BeautifulSoup
```

```python
# specify url
url = "https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue"

# creat request variable
r = requests.get(url)
```

```python
# make contact with website
if r.status_code == 200: 
    html_doc = r.text
```

```python
# Create a Beautiful Soup object

wiki_soup = BeautifulSoup(html_doc)

print(wiki_soup.prettify())
```

```python
# find table elements
table = wiki_soup.find_all('table')

# print all tables 
table
```

```python
# Create a variable and specify the target table as first table
list_companies = table[0]

list_companies
```

```python
# create a variable for all the rows in table 1 
rows = list_companies.find_all('tr')

# storage for extracted data
output=[]

column_names = ["Name", "Industry", "Revenue", "Profit",
                "Employees", "Headquarters", "Ref"]

# create a loop statement 
for x in rows:
    company = x.find_all('td')
    if company:
        company_data = [td.text for td in company]
        output.append(dict(zip(column_names, company_data)))
        
output
```

```python
# view as DataFrame
import pandas as pd 

data = pd.DataFrame(output)

# subset the data to only relevant columns 
data_subset = data[['Name', 'Revenue', 'Headquarters']]
# view filtered table 
data_subset
```
